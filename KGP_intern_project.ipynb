```python
df.Depression.unique()
```

```python
df.columns
```

```python
df_new=df.drop(['Depression','Panic attack','Anxiety','Specialist treatment'],axis='columns')
```

```python
df_new
```

```python
x=df_new
y=df['Depression']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test =train_test_split(x,y, test_size=0.2)
```

```python
from xgboost import XGBClassifier
model= XGBClassifier()
```

```python
model.fit(x_train,y_train)
```

```python
model.predict(x_test)
```

```python
model.score(x_test,y_test)
```

Obtained a AUC-ROC score of 0.839 using XGBoost
